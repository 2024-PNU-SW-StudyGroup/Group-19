```python
import matplotlib.pyplot as plt

# Data for the pie chart
categories = ["DD", "ID", "II", "no RESULT"]
values = [20, 20, 25, 35]

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.title("140")
plt.show()
```

```python
# Data for the two groups: 140분반 and 한국인 일반인
data = {
    "140": [20, 20, 25, 35],  # DD, ID, II, 결과X
    "Korean General Population": [17.5, 46.4, 36.1]  # DD, ID, II
}

categories_140 = ["DD", "ID", "II", "NO RESULT"]
categories_korea = ["DD", "ID", "II"]

# Plot for 140분반
plt.figure(figsize=(12, 6))

# 140분반 원형 그래프
plt.subplot(1, 2, 1)
plt.pie(data["140"], labels=categories_140, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.title("140")

# 한국인 일반인 원형 그래프
plt.subplot(1, 2, 2)
plt.pie(data["Korean General Population"], labels=categories_korea, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.title("Korean General Population")

plt.tight_layout()
plt.show()
```
